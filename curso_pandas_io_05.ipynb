```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect
#create_engine - cria a conexão com o banco de dados
#MetaData - classe que representa os metadados do banco de dados (linhas, colunas, etc)
#Table - classe que representa uma tabela em um banco de dados e é usada para manipular dados em uma tabela específica
#inspect - função que permite inspecionar a estrutura do banco de dados
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
import pandas as pd
```

```python
dados = pd.read_csv(url)
dados.head()
```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"' #Faz uma consulta dentro de uma tabela
#As palavras em caixa alta são as clausulas da pesquisa
```

```python
empregados = pd.read_sql(query, engine)
empregados
```

```python
empregados.to_sql('empregados', con=engine, index=False)
```

```python
pd.read_sql_table('empregados', engine) #Faz a leitura de uma tabela completa
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn: #Garante que a conexão com o banco de dados seja fechada corretamente após a execução da consulta
    conn.execute(query) #Executa a instrução SQL diretamente no banco de dados, usando a conexão estabelecida
```

```python
pd.read_sql_table('clientes', engine) #Podemos verificar que a linha do cliente 5008804 foi excluída
```

```python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(query)
```

```python
pd.read_sql_table('clientes', engine) #Podemos verificar que a escolaridade do cliente 5008808 foi atualizada para "Ensino superior"
```
